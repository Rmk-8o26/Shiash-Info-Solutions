```python
#here we are applying log function
st_log=np.log(hcl)
#it cal the log of an array hcl
plt.plot(st_log)
```

```python
#differnating the log function
plt.figure(figsize=(15,15))
st_log_diff=st_log-st_log.shift()
#the logarithm of the stock and the logarithm of the previous day’s stock 
plt.plot(st_log_diff)
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
model=ARIMA(st_log_diff,order=(2,1,2))
results_arima=model.fit()
plt.plot(st_log_diff)
plt.plot(results_arima.fittedvalues,color='black')
```

```python
import pmdarima as pm
```

```python
def arimamodel(timeseries):
    automodel=pm.auto_arima(timeseries,
                             start_p=2,
                             start_q=2,
                              max_p=10,
                              max_q=10,
                              test='adf',
                              seasonal=True,
                              trace=True)
    return automodel
```

```python
arimamodel(st_log)
```

```python
model=ARIMA(st_log_diff,order=(0,1,0))
results_arima=model.fit()
plt.plot(st_log_diff)
plt.plot(results_arima.fittedvalues,color='black')
```

```python
st_log_diff.fillna(0, inplace=True)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,mean_absolute_percentage_error,r2_score
mae=mean_absolute_error(st_log_diff,results_arima.fittedvalues)
mse=mean_squared_error(st_log_diff,results_arima.fittedvalues)
rmse=np.sqrt(mse)
mape=mean_absolute_percentage_error(st_log_diff,results_arima.fittedvalues)
r2=r2_score(st_log_diff,results_arima.fittedvalues)
```
