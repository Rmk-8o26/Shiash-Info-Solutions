```python
mean_expenditure_by_zone = df.groupby('zone')['expenditure'].mean()
max_expenditure_zone = mean_expenditure_by_zone.idxmax()
mean_expenditure_by_zone,max_expenditure_zone

fig, ax = plt.subplots()
ax.pie(mean_expenditure_by_zone, labels=mean_expenditure_by_zone.index, autopct='%1.1f%%')
ax.axis('equal')  
ax.set_title('Mean Expenditure by Zone')
```

```python
correlation_matrix = df.corr(numeric_only=True)
correlation_matrix
```
